```python
df = pd.read_csv(os.path.join(os.path.expanduser('~/Influence_Scores/score_results'), 'sexist_data_final_train.csv'))
df.head()
```

```python
df = df[['id', 'text', 'numeric_labels']].reset_index(drop = True)
```

```python
run_num = 1
for random_state in (12, 42, 72, 123, 250): # These are the random seeds
    for prune_rate in (5, 10, 15, 20, 25, 30, 35, 40, 50, 60):
        get_random_samples(run_num = run_num, random_state = random_state, prune_rate = prune_rate, data = df)
    run_num  += 1
```
