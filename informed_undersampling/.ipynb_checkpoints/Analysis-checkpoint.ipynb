```python
import pandas as pd
import numpy as np
import os
import math
```

```python
df = pd.read_csv(os.path.join(os.path.expanduser('~/Influence_Scores/score_results'), 'sexist_data_final_train.csv'))
```

```python
df.head()
```

```python
df_pvi = df[['id', 'text', 'numeric_labels', 'average_pvi', 'misclassification_number']].reset_index(drop = True)
```

```python
df_el2n = df[['id', 'average_el2n', 'numeric_labels']].reset_index(drop = True)
```

```python
df_pvi = df_pvi.sort_values(by = 'average_pvi').reset_index(drop = True)
```

```python
df_el2n = df_el2n.sort_values(by = 'average_el2n', ascending = False).reset_index(drop = True)
```

```python
df_pvi_ns = df_pvi[df_pvi['numeric_labels'] == 0].reset_index(drop = True)
df_el2n_ns = df_el2n[df_el2n['numeric_labels'] == 0].reset_index(drop = True)
```

```python
def get_easy_hard_undersample_data(df_ns):
    df_ns_hard = df_ns.iloc[0: math.ceil(0.6 * df_ns.shape[0])].reset_index(drop = True)
    df_ns_easy = df_ns.iloc[-math.ceil(0.6 * df_ns.shape[0]) : -1].reset_index(drop = True)
    df_ns_hard.to_csv(os.path.join('pruned_shots', f'data_pvi_el2n_prune_hard_roberta.csv'), index = False)
    df_ns_easy.to_csv(os.path.join('pruned_shots', f'data_pvi_el2n_prune_easy_roberta.csv'), index = False)
```

```python
df_pvi_el2n_ns = df_pvi_ns.merge(df_el2n_ns[['id', 'average_el2n']].reset_index(drop = True), on = 'id', how = 'inner')
```

```python
get_easy_hard_undersample_data(df_pvi_el2n_ns)
```
