```python
import pandas as pd
import re
import emoji
```

```python
df_sem_eval = pd.read_csv('edos_labelled_aggregated.csv')
```

```python
df_sem_eval
```

```python
def get_clean_text(x):
    x = re.sub('\[USER\]*\s*', '', x).lstrip().rstrip()
    x = re.sub('\[URL\]*\s*', '', x).lstrip().rstrip()
    x = re.sub('\[USER\]*\s*\[URL\]*', '', x).lstrip().rstrip()
    return x
```

```python
def process_emojis(x):
    x = emoji.demojize(x)
    return x
```

```python
df_sem_eval['text'] = df_sem_eval['text'].apply(lambda x: get_clean_text(x))
```

```python
df_sem_eval['text'] = df_sem_eval['text'].apply(lambda x: process_emojis(x))
```

```python
df_sem_eval.to_csv('sem_eval_task10_cleaned.csv', index = False)
```
